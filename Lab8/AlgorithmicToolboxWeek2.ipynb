##2-1: Fibonacci Number 

```python
def calc_fib(n):
  res = [0,1]
  for i in range(2, n+1):
    res.append(res[i-1]+res[i-2])
  return res[n]

n = int(input())
print(calc_fib(n))
```

##2-4: Least Common Multiple

```python
#!/usr/bin/env ruby

def gcd(a, b)
  return a if b == 0;
  gcd(b, (a % b))
end

def lcm(a, b)
  return (a*b)/gcd(a,b)
end

if __FILE__ == $0
  a, b = gets.split().map(&:to_i)
  puts "#{lcm(a, b)}"
end
```

##2-8: Last Digit of the Sum of Squares of Fibonacci Numbers

```python
# Uses python3
from sys import stdin

def calculteNextFibNumbers(fibNumbers):
  temp = fibNumbers[0]
  fibNumbers[0] = fibNumbers[1]
  fibNumbers[1] = temp + fibNumbers[0]
  return fibNumbers

def getFib(n):
  if n == 0:
    return 0 
  fibNumbers = [0,1]
  for i in range(1,n):
    fibNumbers = calculteNextFibNumbers(fibNumbers)
  return fibNumbers[1]

def fibonacci_sum_squares_naive(n):
    BASE = 10
    PISANO_PERIOD = 60
    n = n%PISANO_PERIOD
    return (getFib(n)*getFib(n+1))%BASE

if __name__ == '__main__':
    n = int(stdin.read())
    print(fibonacci_sum_squares_naive(n))
```
